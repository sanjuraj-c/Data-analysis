```python
data_cleaned=df.dropna(axis=1)
print(data_cleaned)
```

```python
column1=df['continent'].fillna(method='ffill')
print(column1)
```

```python
column2=df['total_cases'].fillna('265748.0')
print(column2)
```

```python
column3 = df['total_deaths'].fillna(0)
print(column3)
```

```python
column4=df['median_age'].fillna('ffill')
print(column4)
```

```python
column5=df['gdp_per_capita'].fillna(0)
print(column5)
```

```python
column6=df['aged_65_older'].fillna('bfill')
print(column6)
```

```python

column7=df['life_expectancy'].fillna(0)
print(column7)
```

```python
column8=df['new_cases'].fillna(0)
print(column8)
```

```python
column9=df['new_deaths'].fillna(0)
print(column9)
```

```python
column10=df['new_tests'].fillna(df['new_tests'].mean())
print(column10)
```

```python
column11=df['total_tests'].fillna(df['total_tests'].mean())
print(column11)
```

```python
column12=df['total_vaccinations'].fillna(df['total_vaccinations'].mean())
print(column12)
```

```python

column13=df['excess_mortality'].fillna(df['excess_mortality'].mean())
print(column13)
```

```python
data_cleaned2=pd.concat([data_cleaned,column1,column2,column3,column4,column5,column6,column7,column8,column9,column10,column11,column12,column13],axis=1)
print(data_cleaned2)
```

```python

data_cleaned2['total_cases'] = pd.to_numeric(data_cleaned2['total_cases'], errors='coerce')
continent_total_cases = data_cleaned2.groupby('continent')['total_cases'].sum()

plt.figure(figsize=(10, 6))
continent_total_cases.plot(kind='bar', color='skyblue')
plt.title('COVID-19 total cases by Continent')
plt.xlabel('Continent')
plt.ylabel('Total covid Cases')
plt.xticks(rotation=45)
plt.show()
```

```python
data_cleaned2['date'] = pd.to_datetime(data_cleaned2['date'])

# Sort the DataFrame by date
data_cleaned2 = data_cleaned2.sort_values(by='date')

# Create the time series plot
plt.figure(figsize=(12, 6))
plt.plot(data_cleaned2['date'], data_cleaned2['new_cases'], label='New Cases', linestyle='-')
plt.title('COVID-19 New Cases Time Series')
plt.xlabel('Date')
plt.ylabel('New Cases')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Filter data for continent=Asia
asia_data = data_cleaned2[data_cleaned2['continent'] == 'Asia']

# Create scatter plot for total_cases vs total_deaths in Asia
plt.figure(figsize=(10, 6))
plt.scatter(asia_data['total_cases'], asia_data['total_vaccinations'], color='orange', alpha=0.7)
plt.title('COVID-19 Total Cases vs Total vaccinations in Asia')
plt.xlabel('Total Cases')
plt.ylabel('Total vaccinations')
plt.grid(True)
plt.show()
```

```python
europe_data = data_cleaned2[data_cleaned2['continent'] == 'Europe']

# Create a scatter plot for total_cases vs total_deaths in Europe
plt.figure(figsize=(10, 6))
plt.scatter(europe_data['total_cases'], europe_data['total_vaccinations'], color='blue', alpha=0.7)
plt.title('COVID-19 Total Cases vs Total vaccinations in Europe')
plt.xlabel('Total Cases')
plt.ylabel('Total vaccinations')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
data_cleaned2.boxplot(column='new_deaths', by='continent', grid=False)
plt.title('COVID-19 Deaths by Continent')
plt.xlabel('Continent')
plt.ylabel('New Deaths')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Plotting new_cases
plt.plot(data_cleaned2['date'], data_cleaned2['new_cases'], color='skyblue', label='New Cases', linewidth=2)

# Plotting new_deaths
plt.plot(data_cleaned2['date'], data_cleaned2['new_deaths'], color='red', label='New Deaths', linewidth=2)

# Adding labels and title
plt.title('COVID-19 New Cases vs New Deaths')
plt.xlabel('Date')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.legend()  # Show legend to distinguish between new_cases and new_deaths
plt.grid(True)

# Display the line chart
plt.tight_layout()  # Automatically adjust subplot params for better layout
plt.show()











```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data_cleaned2['life_expectancy'], color='skyblue')
plt.title('Life Expectancy')
plt.xlabel('Life Expectancy')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Create a line plot for excess_mortality
plt.plot(data_cleaned2['date'], data_cleaned2['excess_mortality'], color='green', marker='o', linestyle='-', linewidth=2, markersize=6)
plt.title('Excess Mortality Trend')
plt.xlabel('Date')
plt.ylabel('Excess Mortality')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Create a histogram for gdp_per_capita
plt.hist(data_cleaned2['gdp_per_capita'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of GDP per Capita')
plt.xlabel('GDP per Capita')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
data_cleaned2['aged_65_older'] = pd.to_numeric(data_cleaned2['aged_65_older'], errors='coerce')
plt.figure(figsize=(10, 6))
plt.plot(data_cleaned2['aged_65_older'], data_cleaned2['total_cases'], color='blue', marker='o')
plt.title('COVID-19 Total Cases vs Aged 65 and Older')
plt.xlabel('Aged 65 and Older (%)')
plt.ylabel('Total Cases')
plt.grid(True)
plt.show()
```

```python
# Replace non-numeric values with NaN in 'total_cases' and 'median_age' columns
data_cleaned2['total_cases'] = pd.to_numeric(data_cleaned2['total_cases'], errors='coerce')
data_cleaned2['median_age'] = pd.to_numeric(data_cleaned2['median_age'], errors='coerce')

# Drop rows with NaN values in 'total_cases' and 'median_age' columns
data_cleaned2 = data_cleaned2.dropna(subset=['total_cases', 'median_age'])

# Now create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(data_cleaned2['total_cases'], data_cleaned2['median_age'], color='orange', alpha=0.7)
plt.title('COVID-19 Total Cases vs Median Age')
plt.xlabel('Total Cases')
plt.ylabel('Median Age')
plt.grid(True)
plt.show()



```

```python
plt.figure(figsize=(10, 6))
plt.plot(data_cleaned2['population'], data_cleaned2['total_vaccinations'], color='green', marker='o')
plt.title('Population vs Total Vaccinations')
plt.xlabel('Population')
plt.ylabel('Total Vaccinations')
plt.grid(True)
plt.show()
```

```python
heatmap_data = data_cleaned2.groupby(['date', 'continent'])['total_cases'].sum().reset_index()

# Pivot the data for heatmap
heatmap_data_pivot = heatmap_data.pivot('date', 'continent', 'total_cases')

# Create a heatmap using seaborn
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data_pivot, cmap='YlGnBu', annot=True, fmt='g', linewidths=.5)
plt.title('COVID-19 Total Cases Heatmap')
plt.xlabel('Continent')
plt.ylabel('Date')
plt.show()
```

```python
plt.figure(figsize=(8, 8))

# Calculate total cases by continent
total_cases_by_continent = data_cleaned2.groupby('continent')['total_tests'].sum()

# Create a pie chart
plt.pie(total_cases_by_continent, labels=total_cases_by_continent.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Distribution of Total tests by Continent')
plt.show()
```
