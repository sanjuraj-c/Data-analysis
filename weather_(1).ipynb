```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
excel_file_path ='Weather data Petrolina Station_1minute_01-01-2015_31-01-2015 (1).xlsx'
sheet_name = 1
df=pd.read_excel(excel_file_path,sheet_name = sheet_name)
df
```

```python
df.drop_duplicates()
```

```python
df.dropna()
```

```python
cl1 = df['glo_avg'].fillna(df['glo_avg'].mean())
print(cl1)
```

```python
cl2 = df['dir_avg'].fillna(df['dir_avg'].mean())
```

```python
cl3 = df['diff_avg'].fillna(0)
```

```python
cl4 = df['lw_avg'].fillna(df['lw_avg'].mean())
```

```python
cl5 = df['par_avg'].fillna(df['par_avg'].mean())
```

```python
cl6 = df['lux_avg'].fillna(df['lux_avg'].mean())
```

```python
cl7 =  df['tp_sfc'].fillna(df['tp_sfc'].mean())
```

```python
cl8 =  df['humid'].fillna(df['humid'].mean())
```

```python
cl9 =  df['press'].fillna(df['press'].mean())
```

```python
cl10 =  df['rain'].fillna(0)
```

```python
cl11 =  df['ws_10m'].fillna(df['ws_10m'].mean())
```

```python
cl12 =  df['wd_10m'].fillna(df['wd_10m'].mean())
```

```python
df1 = pd.concat([df['id'],df['day'],df['year'],df['min'],cl1,cl2,cl3,cl4,cl5,cl6,cl7,cl8,cl9,cl10,cl11,cl12],axis=1)
df1
```

```python
df1.isnull().sum()
```

```python
plt.figure(figsize=(8, 6))
sns.lineplot(x='day', y='tp_sfc', data=df1, marker='o')
plt.xlabel('Day')
plt.ylabel('Temperature (tp_sfc)')
plt.title('Temperature Variation over Days')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='rain', y='humid', data=df1)
plt.xlabel('Rain')
plt.ylabel('Humidity')
plt.title('Humidity Distribution by Rain Status')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='year', y='glo_avg', data=df1, ci=None)
plt.xlabel('Year')
plt.ylabel('Average Global Radiation')
plt.title('Average Global Radiation by Year')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='lw_avg', y='press', data=df1, hue='year')
plt.xlabel('long wave radiation')
plt.ylabel('atm pressure')
plt.title('long wave rad vs. atmos press')
plt.show()
```

```python
sns.countplot(df['rain'])
plt.title('rain')
```

```python

selected_columns = ['glo_avg', 'tp_sfc', 'rain','humid','press']

# Calculate the sum of values for each selected column
column_sums = df1[selected_columns].sum()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(column_sums, labels=column_sums.index, autopct='%1.1f%%', startangle=140)
plt.title('Weather Report Data Distribution')
plt.show()
```

```python
columns = [
    'min', 'glo_avg', 'dir_avg', 'diff_avg', 'lw_avg',
    'par_avg', 'lux_avg', 'tp_sfc', 'humid', 'press', 'rain',
    'ws_10m', 'wd_10m'
]

data = df[columns]

# Compute the correlation matrix
corr_matrix = data.corr()

# Create a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()

```
